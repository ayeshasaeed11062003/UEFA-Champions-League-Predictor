```python
import requests
```

URL for Champions League Statistics for the 2024-25 season (currently ongoing)

```python
standings_url = "https://fbref.com/en/comps/8/Champions-League-Stats"
```

```python
data = requests.get(standings_url)
```

Making our standings_table from the URLs in the League Table. Beautiful Soup to parse HTML, extracting the text from our 'data' and specifying our required table within the table class (stats_table), since we only want the first table to scrape, adding [0]. Table name (stats_table) taken from developer tools. 'select' method uses CSS selectors

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(data.text)
standings_table = soup.select('table.stats_table')[0]
```

Inside the table, selecting all the anchor tags (a tags). Using findall instead of select here since we only want tags.

```python
links = standings_table.find_all('a')
links = [l.get("href") for l in links] # grabbing the 'href' property from all 'a' elements
links = [l for l in links if '/squads/' in l] # only get the links with '/squad/' in them
links
```

As we can see, this links only contain the last part of the URL, so we'll use format string to complete the URLs

```python
team_urls = [f"https://fbref.com{l}" for l in links]
```

```python
team_urls
```

Now, that we have full links, we can analyse stats for specific teams. For our project, we'll start with looking at the top team right now: Liverpool

```python
liverpool_url = team_urls[0]
liverpool_data = requests.get(liverpool_url)
```

Now, we want to analyse Liverpool's play. For this we're going to get the Scores and Fixtures table for liverpool and turn it into a dataframe. The 'match' function is going to look for the 'Scores and Fixtures' string inside that table, whereas read_html method reads all the tableson the page. This way, we can read all the pages and select the one captioned 'Scores & Fixtures'. Alternatively we can use the id for the table to match it with, but we're using the caption tag here.

```python
import pandas as pd
liverpool_matches = pd.read_html(liverpool_data.text, match="Scores & Fixtures")
```

```python
liverpool_matches
```

This resulted in a list, so now let's take the first element of the list to give us a dataframe

Now we have just the basic stats like goals for, goals against, attendance, captain etc. but no detailed stats, so we'll dive deeper. There are detailed tables on shooting, goalkeeping, passing, pass types, goal and shot creation, defensive actions, posession and miscellaneous stats. So let's look at shooting stats

```python
soup2 = BeautifulSoup(liverpool_data.text)
```

Now since all the tables mentioned above are referenced by a tags, we'll use findall again.

```python
more_links = soup2.find_all('a') # finding all anchor (a) tabs
more_links = [l.get("href") for l in more_links] # getting the href property again
more_links = [l for l in more_links if l and 'all_comps/shooting' in l] # using 'if l' here beacuse some links don't have href property
```

like before, we're completing the URLs and matching on the string 'Shooting'

```python
liverpool_data = requests.get(f"https://fbref.com{more_links[0]}")
liverpool_shooting = pd.read_html(liverpool_data.text, match="Shooting")[0]
liverpool_shooting.head()
```

now to remove the multi level index, dropping one of them (the top one) since it is unnecessary

```python
liverpool_shooting.columns = liverpool_shooting.columns.droplevel()
```

```python
liverpool_shooting.head()
```

Now, we have a liverpool_matches dataframe and a liverpool_shooting dataframe, but we want to combine it. We can do this because they're referencing the same matches but with varied stats. Using liverpool nickname 'reds' for naming now. Selecting relevant rows from the liverpool_shooting table (since many rows are duplicative) such as Date, Sh (total shots excluding penalties), SoT (shots on target), Dist (avg distance travelled by a shot), FK (free kicks), PK (penalty kicks), PKatt (penalty kicks attempted)

```python
# merging on date column (since its in both tables)
reds_data = liverpool_matches[0].merge(liverpool_shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date") 
```
